```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("company_data_sales.csv")
```

### Total profit of all months using a line plot


```python
X=df['Month Number'].to_list()
y=df['Total Profit'].to_list()
```

```python
plt.plot(X,y)
plt.xticks(X)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company profit per Month")
plt.show()
```

### Total profit of all months using line plot


```python
plt.plot(X,y,'o:r',linewidth=3,label='Profit data of last year')
plt.xticks(X)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company profit per Month")
plt.legend(loc=4)
plt.show()
```

### All product sales data using a multiline plot

```python
facecream=df['Facecream'].to_list()
facewash=df['Facewash'].to_list()
toothpaste=df['Toothpaste'].to_list()
bathingsoap=df['Bathing Soap'].to_list()
shampoo=df['Shampoo'].to_list()
moisturizer=df['Moisturizer'].to_list()
X=df['Month Number'].to_list()
```

```python
plt.plot(X,facecream,'o-',linewidth=3,label="Face Cream")
plt.plot(X,facewash,'o-',linewidth=3,label="Face Wash")
plt.plot(X,toothpaste,'o-',linewidth=3,label="Toothpaste")
plt.plot(X,bathingsoap,'o-',linewidth=3,label="Bathing Soap")
plt.plot(X,shampoo,'o-',linewidth=3,label="Shampoo")
plt.plot(X,moisturizer,'o-',linewidth=3,label="Moisturizer")
plt.xticks(X)
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Sales Data")
plt.legend()
plt.show()
```

### The total profit of each month and using the histogram to see the most common profit ranges

### Total sale data for last year for each product and using a Pie chart

```python
facecream=df['Facecream']
facewash=df['Facewash']
toothpaste=df['Toothpaste']
bathingsoap=df['Bathing Soap']
shampoo=df['Shampoo']
moisturizer=df['Moisturizer']
X=df['Month Number']
```

```python
sumdata=[facecream.sum(),facewash.sum(),toothpaste.sum(),bathingsoap.sum(),shampoo.sum(),moisturizer.sum()]
```

```python
labels=["Facecream","Facewash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"]
plt.pie(sumdata,labels=labels,autopct='%1.1f%%')
plt.legend(loc=10)
plt.show()
```

### Bathing soap facewash of all months using the Subplot

```python
X=df['Month Number'].to_list()
bathingsoap=df['Bathing Soap'].to_list()
facewash=df['Facewash'].to_list()
```

```python
plt.subplot(2,1,1)
plt.plot(X,bathingsoap,"o-r",linewidth=3)
plt.title("Sales data of bathingsoap")
plt.xticks(X)

plt.subplot(2,1,2)
plt.plot(X,facewash,"o-k",linewidth=3)
plt.title("Sales data of facewash")
plt.xlabel("Month Number")
plt.ylabel("Sales in unit number")
plt.xticks(X)

plt.tight_layout()
plt.show()
```

### All product sales data using the stack plot

```python
facecream=df['Facecream'].to_list()
facewash=df['Facewash'].to_list()
toothpaste=df['Toothpaste'].to_list()
bathingsoap=df['Bathing Soap'].to_list()
shampoo=df['Shampoo'].to_list()
moisturizer=df['Moisturizer'].to_list()
X=df['Month Number'].to_list()
```

```python
labels=["Facecream","Facewash","Toothpaste","Bathingsoap","Shampoo","Moisturizer"]
plt.stackplot(X,facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer,labels=labels)
plt.legend()
plt.xlabel("Month Number")
plt.ylabel("Sales unit in Number")
plt.title("All product sales using stack plot")
plt.show()
```
